# **AI-ML Model - K Means**

```python
import matplotlib.pyplot as mp
%matplotlib inline
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
from sklearn.cluster import KMeans
```

```python
df
```

***Checking Null Values***

```python
pd.isnull(df).sum()
```

```python
df.describe()
```

```python
kmeans = KMeans(n_clusters = 4)
```

```python
kmeans.fit(df)
```

```python
pred = kmeans.predict(df)
```

```python
pred
```

```python
pd.Series(pred).value_counts()
```

```python
kmeans.inertia_
```

```python
kmeans.score(df)
```

```python
SSE = []
```

```python
for cluster in range(1, 20):
    kmeans = KMeans(n_jobs = -1, n_clusters = cluster)
    kmeans.fit(df)
    SSE.append(kmeans.inertia_)
```

```python
frame = pd.DataFrame({'Cluster': range(1, 20), 'SSE': SSE})
```

```python
frame.head()
```

```python
mp.figure(figsize = (12, 6))
mp.plot(frame['Cluster'], frame['SSE'], marker = "o")
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
df_scaled = scaler.fit_transform(df)
```

```python
pd.DataFrame(df_scaled).describe()
```

```python
SSE_scaled = []
```

```python
for cluster in range(1, 20):
    kmeans = KMeans(n_jobs = -1, n_clusters = cluster)
    kmeans.fit(df_scaled)
    SSE_scaled.append(kmeans.inertia_)
```

```python
frame_scaled = pd.DataFrame({'Cluster': range(1, 20), 'SSE': SSE_scaled})
```

```python
mp.plot(frame_scaled['Cluster'], frame_scaled['SSE'], marker = "o")
mp.xlabel("Clusters")
mp.ylabel("SSE")
```

```python
kmeans = KMeans(n_jobs = -1, n_clusters = 6)
```

```python
kmeans.fit(df_scaled)
```

```python
pred = kmeans.predict(df_scaled)
```

```python
frame = pd.DataFrame(df_scaled)
```
